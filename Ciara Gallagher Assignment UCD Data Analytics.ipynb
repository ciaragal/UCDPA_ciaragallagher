```python
# I then created a custom function to create my graphs.
```

```python
sns.countplot(data=mmfs_data, x='Issuer_Region', hue='Holder_Sector')
plt.xticks(rotation=90)
plt.legend(loc='upper right')
plt.xlabel('Issuer Region')
plt.ylabel('Millions (euro)')
plt.title('Money Market Fund Shares by location, holder sector and amount')
```

```python
sns.countplot(data=dslt_data, x='Issuer_Region', hue='Holder_Sector')
plt.xticks(rotation=90)
plt.legend(loc='upper right')
plt.xlabel('Issuer Region')
plt.ylabel('Millions (euro)')
plt.title('Debt Securities - Long-Term by location, holder sector and amount')
```
